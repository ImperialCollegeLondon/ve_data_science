
# Individual Variable Analysis Workflow

## Manual Single-Variable Processing

The following workflow processes and visualises
a single hydrological variable at a time.

The target variable is manually edited:

e.g., variable = "subsurface_flow"


The workflow then generates:

* variable metadata
* summary statistics
* spatial distribution figures
* temporal variability figures
* suggested structure of interpretation notes

for the selected variable only.

---

# Advantages of Individual Mode

Using individual-variable mode provides:

| Advantage                     | Description                                            |
| ----------------------------- | ------------------------------------------------------ |
| Flexible interpretation       | Manual markdown notes can be added after each variable |
| Easier manuscript preparation | Figures and interpretation can be refined individually |
| Interactive analysis          | Variables can be explored one-by-one                   |
| Better figure customisation   | Plot settings can be adjusted per variable             |
| Easier debugging              | Issues can be isolated to specific variables           |
| Reduced notebook clutter      | Only selected outputs are generated                    |

# Limitations

Although individual-variable workflows are flexible,
there are several limitations:

| Limitation                   | Description                               |
| ---------------------------- | ----------------------------------------- |
| Repetitive manual execution  | Variables must be changed manually        |
| Slower for many variables    | Each variable requires separate execution |
| Less suitable for automation | Manual interaction is required            |
| Inconsistent formatting risk | Manual edits may reduce standardisation   |




# Surface Runoff

## Purpose

This section investigates:
- runoff generation
- hydrological connectivity
- spatial runoff organisation
- temporal runoff variability

## Expected Behaviour

Surface runoff is expected to:
- increase during precipitation events
- respond rapidly to rainfall forcing
- highlight hydrologically connected areas
- show spatial runoff accumulation patterns

`````python
# =====================================================
# SELECT VARIABLE
# =====================================================

variable = "surface_runoff"

# =====================================================
# READ VARIABLE METADATA
# =====================================================

metadata = flow_variables.get(variable, {})

unit = metadata.get("unit", "")

description = metadata.get("description", "")

# =====================================================
# CATEGORY
# =====================================================

category = classify_variable(ds[variable])

# =====================================================
# VARIABLE DATA
# =====================================================

da = ds[variable]

# =====================================================
# VARIABLE HEADER
# =====================================================

display(
    Markdown(
        f"""
# {variable}

---

## Variable Information

| Property | Value |
|---|---|
| Variable | {variable} |
| Category | {category} |
| Unit | {unit} |
| Description | {description} |
| Dimensions | {da.dims} |
| Shape | {da.shape} |

---
"""
    )
)


# =====================================================
# SUMMARY STATISTICS
# =====================================================

display(Markdown("## Summary Statistics"))

summary = create_summary_table(ds, variable)

display(summary)

# =====================================================
# QUICK SUMMARY
# =====================================================

print_variable_summary(ds, variable)

# =====================================================
# SCALAR VARIABLES
# =====================================================

if category == "scalar":
    # -------------------------------------------------
    # SPATIAL MAP
    # -------------------------------------------------

    display(Markdown("## Mean Spatial Distribution Across All Timesteps"))

    # =================================================
    # COMPUTE TEMPORAL MEAN
    # =================================================

    da_mean = ds[variable].mean(dim="time_index")

    # =================================================
    # RESHAPE CELL_ID -> 2D GRID
    # =================================================

    nx = len(np.unique(ds["x"].values))

    ny = len(np.unique(ds["y"].values))

    grid = da_mean.values.reshape(ny, nx)

    # =================================================
    # CREATE FIGURE
    # =================================================

    fig, ax = plt.subplots(figsize=(7, 6))

    # =================================================
    # SPATIAL PLOT
    # =================================================

    im = ax.imshow(
        grid,
        origin="lower",
        extent=[
            ds["x"].values.min(),
            ds["x"].values.max(),
            ds["y"].values.min(),
            ds["y"].values.max(),
        ],
        aspect="auto",
    )

    # =================================================
    # COLORBAR
    # =================================================

    cbar = plt.colorbar(im, ax=ax)

    cbar.set_label(f"{variable} ({unit})")

    # =================================================
    # AXIS LABELS
    # =================================================

    ax.set_xlabel(
        "x (UTM50N)",
        fontsize=11,
        labelpad=10,
    )

    ax.set_ylabel(
        "y (UTM50N)",
        fontsize=11,
        labelpad=10,
    )

    # =================================================
    # PREVENT LABEL OVERLAP
    # =================================================

    ax.tick_params(
        axis="x",
        rotation=45,
        labelsize=9,
    )

    ax.tick_params(
        axis="y",
        labelsize=9,
    )

    ax.locator_params(
        axis="x",
        nbins=6,
    )

    ax.locator_params(
        axis="y",
        nbins=6,
    )

    # =================================================
    # LAYOUT
    # =================================================

    plt.tight_layout()

    plt.show()

    # -------------------------------------------------
    # TEMPORAL VARIABILITY
    # -------------------------------------------------

    display(Markdown("## Temporal Variability Across Grid Cells"))

    plot_scalar_timeseries(ds, variable=variable, unit=unit)

    plt.show()

# =====================================================
# LAYERED VARIABLES
# =====================================================

elif category == "layered":
    # -------------------------------------------------
    # LAYER MAP
    # -------------------------------------------------

    display(Markdown("## Layered Spatial Distribution"))

    plot_layered_map(ds, variable=variable, layer=0, timestep=0, unit=unit)

    plt.show()

    # -------------------------------------------------
    # TIMESERIES
    # -------------------------------------------------

    display(Markdown("## Layered Temporal Variability"))

    plot_layered_timeseries(ds, variable=variable, layer=0, unit=unit)

    plt.show()


# =====================================================
# INTERPRETATION REMINDER
# =====================================================

display(
    Markdown(
        f"""
> Add interpretation markdown cell below for:
>
> **{variable}**
>
> Suggested topics:
> - observations
> - spatial behaviour
> - temporal behaviour
> - hydrological implications
> - follow-up analyses
"""
    )
)
`````


# Automated Loop Workflow 

The following workflow runs in **loop mode** and automatically
processes all variables listed in:

```python
flow_variables
````

during a single execution.

This means the notebook will automatically:

* iterate through every hydrological flow variable
* generate summary statistics
* create spatial distribution figures
* create temporal variability figures
* generate markdown interpretation sections

without manually changing:

variable = "..."

for each run.

---

# Advantages of Loop Mode

Using loop mode provides several benefits:

* avoids repetitive manual analysis
* standardises all figures
* ensures consistent formatting
* speeds up exploratory analysis
* supports automated reporting

# Limitations

Although loop mode greatly improves automation,
there are several limitations:

| Limitation                                        | Description                                                                                                                                               |
| ------------------------------------------------- | --------------------------------------------------------------------------------------------------------------------------------------------------------- |
| No manual markdown insertion between variables    | Because all variables run continuously inside a loop, manual markdown cells cannot be inserted interactively after each variable section during execution |
| Reduced flexibility for individual figure editing | All variables follow the same workflow structure and formatting                                                                                           |
| Large notebook outputs                            | Simulations with many variables may produce very long notebook outputs                                                                                    |
| Higher memory usage                               | Large datasets and many figures may increase RAM consumption                                                                                              |
| Longer execution time                             | Full automated workflows may take several minutes depending on grid size and simulation duration                                                          |
| Difficult interactive interpretation              | Real-time discussion and interpretation after each variable is less convenient in loop mode                                                               |

`````python
# =====================================================
# LOOP THROUGH ALL FLOW VARIABLES
# =====================================================

for variable in flow_variables.keys():
    # =================================================
    # READ VARIABLE METADATA
    # =================================================

    metadata = flow_variables.get(variable, {})

    unit = metadata.get("unit", "")

    description = metadata.get("description", "")

    # =================================================
    # CATEGORY
    # =================================================

    category = classify_variable(ds[variable])

    # =================================================
    # VARIABLE DATA
    # =================================================

    da = ds[variable]

    # =================================================
    # VARIABLE HEADER
    # =================================================

    display(
        Markdown(
            f"""
# {variable}

---

## Variable Information

| Property | Value |
|---|---|
| Variable | {variable} |
| Category | {category} |
| Unit | {unit} |
| Description | {description} |
| Dimensions | {da.dims} |
| Shape | {da.shape} |

---
"""
        )
    )

    # =================================================
    # SUMMARY STATISTICS
    # =================================================

    display(Markdown("## Summary Statistics"))

    summary = create_summary_table(ds, variable)

    display(summary)

    # =================================================
    # QUICK SUMMARY
    # =================================================

    print_variable_summary(ds, variable)

    # =================================================
    # SCALAR VARIABLES
    # =================================================

    if category == "scalar":
        # -------------------------------------------------
        # SPATIAL MAP
        # -------------------------------------------------

        display(Markdown("## Mean Spatial Distribution Across All Timesteps"))

        # =============================================
        # TEMPORAL MEAN
        # =============================================

        da_mean = ds[variable].mean(dim="time_index")

        # =============================================
        # RESHAPE GRID
        # =============================================

        nx = len(np.unique(ds["x"].values))

        ny = len(np.unique(ds["y"].values))

        grid = da_mean.values.reshape(ny, nx)

        # =============================================
        # CREATE FIGURE
        # =============================================

        fig, ax = plt.subplots(figsize=(7, 6))

        # =============================================
        # SPATIAL PLOT
        # =============================================

        im = ax.imshow(
            grid,
            origin="lower",
            extent=[
                ds["x"].values.min(),
                ds["x"].values.max(),
                ds["y"].values.min(),
                ds["y"].values.max(),
            ],
            aspect="auto",
        )

        # =============================================
        # COLORBAR
        # =============================================

        cbar = plt.colorbar(im, ax=ax)

        cbar.set_label(f"{variable} ({unit})")

        # =============================================
        # AXIS LABELS
        # =============================================

        ax.set_xlabel("x (UTM50N)")

        ax.set_ylabel("y (UTM50N)")

        plt.tight_layout()

        plt.show()

        # -------------------------------------------------
        # TEMPORAL VARIABILITY
        # -------------------------------------------------

        display(Markdown("## Temporal Variability Across Grid Cells"))

        plot_scalar_timeseries(
            ds,
            variable=variable,
            unit=unit,
        )

        plt.show()

    # =================================================
    # LAYERED VARIABLES
    # =================================================

    elif category == "layered":
        # -------------------------------------------------
        # LAYER MAP
        # -------------------------------------------------

        display(Markdown("## Layered Spatial Distribution"))

        plot_layered_map(
            ds,
            variable=variable,
            layer=0,
            timestep=0,
            unit=unit,
        )

        plt.show()

        # -------------------------------------------------
        # TIMESERIES
        # -------------------------------------------------

        display(Markdown("## Layered Temporal Variability"))

        plot_layered_timeseries(
            ds,
            variable=variable,
            layer=0,
            unit=unit,
        )

        plt.show()

    # =================================================
    # INTERPRETATION REMINDER
    # =================================================

    display(
        Markdown(
            f"""
> Add interpretation markdown cell below for:
>
> **{variable}**
>
> Suggested topics:
> - observations
> - spatial behaviour
> - temporal behaviour
> - hydrological implications
> - follow-up analyses
"""
        )
    )

    # =================================================
    # SECTION SEPARATOR
    # =================================================

    display(Markdown("---\n---\n---"))
`````
