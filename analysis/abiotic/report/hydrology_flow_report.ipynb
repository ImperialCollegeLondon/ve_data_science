# Surface Runoff

## Purpose

This section investigates:
- runoff generation
- hydrological connectivity
- spatial runoff organisation
- temporal runoff variability

## Expected Behaviour

Surface runoff is expected to:
- increase during precipitation events
- respond rapidly to rainfall forcing
- highlight hydrologically connected areas
- show spatial runoff accumulation patterns

```python
#=====================================================
# SELECT VARIABLE
# =====================================================

variable = "surface_runoff"

# =====================================================
# READ VARIABLE METADATA
# =====================================================

metadata = flow_variables.get(
    variable,
    {}
)

unit = metadata.get(
    "unit",
    ""
)

description = metadata.get(
    "description",
    ""
)

# =====================================================
# CATEGORY
# =====================================================

category = classify_variable(
    ds[variable]
)

# =====================================================
# VARIABLE DATA
# =====================================================

da = ds[variable]

# =====================================================
# VARIABLE HEADER
# =====================================================

display(
    Markdown(
        f"""
# {variable}

---

## Variable Information

| Property | Value |
|---|---|
| Variable | {variable} |
| Category | {category} |
| Unit | {unit} |
| Description | {description} |
| Dimensions | {da.dims} |
| Shape | {da.shape} |

---
"""
    )
)


# =====================================================
# SUMMARY STATISTICS
# =====================================================

display(
    Markdown(
        "## Summary Statistics"
    )
)

summary = create_summary_table(
    ds,
    variable
)

display(summary)

# =====================================================
# QUICK SUMMARY
# =====================================================

print_variable_summary(
    ds,
    variable
)

# =====================================================
# SCALAR VARIABLES
# =====================================================

if category == "scalar":

    # -------------------------------------------------
    # SPATIAL MAP
    # -------------------------------------------------

    display(
        Markdown(
            "## Mean Spatial Distribution Across All Timesteps"
        )
    )

    # =================================================
    # COMPUTE TEMPORAL MEAN
    # =================================================

    da_mean = ds[variable].mean(
        dim="time_index"
    )

    # =================================================
    # RESHAPE CELL_ID -> 2D GRID
    # =================================================

    nx = len(np.unique(ds["x"].values))

    ny = len(np.unique(ds["y"].values))

    grid = da_mean.values.reshape(
        ny,
        nx
    )

    # =================================================
    # CREATE FIGURE
    # =================================================

    fig, ax = plt.subplots(
        figsize=(7, 6)
    )

    # =================================================
    # SPATIAL PLOT
    # =================================================

    im = ax.imshow(
        grid,
        origin="lower",
        extent=[
            ds["x"].values.min(),
            ds["x"].values.max(),
            ds["y"].values.min(),
            ds["y"].values.max(),
        ],
        aspect="auto",
    )

    # =================================================
    # COLORBAR
    # =================================================

    cbar = plt.colorbar(
        im,
        ax=ax
    )

    cbar.set_label(
        f"{variable} ({unit})"
    )


    # =================================================
    # AXIS LABELS
    # =================================================

    ax.set_xlabel(
        "x (UTM50N)",
        fontsize=11,
        labelpad=10,
    )

    ax.set_ylabel(
        "y (UTM50N)",
        fontsize=11,
        labelpad=10,
    )

    # =================================================
    # PREVENT LABEL OVERLAP
    # =================================================

    ax.tick_params(
        axis="x",
        rotation=45,
        labelsize=9,
    )

    ax.tick_params(
        axis="y",
        labelsize=9,
    )

    ax.locator_params(
        axis="x",
        nbins=6,
    )

    ax.locator_params(
        axis="y",
        nbins=6,
    )

    # =================================================
    # LAYOUT
    # =================================================

    plt.tight_layout()

    plt.show()
    
    # -------------------------------------------------
    # TEMPORAL VARIABILITY
    # -------------------------------------------------

    display(
        Markdown(
            "## Temporal Variability Across Grid Cells"
        )
    )

    plot_scalar_timeseries(
        ds,
        variable=variable,
        unit=unit
    )

    plt.show()

# =====================================================
# LAYERED VARIABLES
# =====================================================

elif category == "layered":

    # -------------------------------------------------
    # LAYER MAP
    # -------------------------------------------------

    display(
        Markdown(
            "## Layered Spatial Distribution"
        )
    )

    plot_layered_map(
        ds,
        variable=variable,
        layer=0,
        timestep=0,
        unit=unit
    )

    plt.show()

    # -------------------------------------------------
    # TIMESERIES
    # -------------------------------------------------

    display(
        Markdown(
            "## Layered Temporal Variability"
        )
    )

    plot_layered_timeseries(
        ds,
        variable=variable,
        layer=0,
        unit=unit
    )

    plt.show()


# =====================================================
# INTERPRETATION REMINDER
# =====================================================

display(
    Markdown(
        f"""
> Add interpretation markdown cell below for:
>
> **{variable}**
>
> Suggested topics:
> - observations
> - spatial behaviour
> - temporal behaviour
> - hydrological implications
> - follow-up analyses
"""
    )
)
```
